# Data Analysis

## Top 5 most popular artist 

```python
top_five_artists = df.groupby("artist").count().sort_values(by="song_title",ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

## Top 5 Loudest Track

```python
top_five_loudest_track = df[["loudness","song_title"]].sort_values(by="loudness",ascending=False)[:5]
top_five_loudest_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y="song_title", data=top_five_loudest_track)
plt.title("top 5 loudess track")
plt.show()
```

## Artist with most danceability songs 

```python
top_five_artist_danceable_songs = df[["danceability","song_title", "artist"]].sort_values(by="danceability",
                                                                                          ascending=False)[:5]
top_five_artist_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability",y="artist", data=top_five_artist_danceable_songs)
plt.title("Artist with most Danceability song")
plt.show()
```

## Top ten instrumental tracks 

```python
top_ten_instrumental_tracks = df[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",
                                                                                        ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_ten_instrumental_tracks,autopct='%1.2f%%',
        labels=top_ten_instrumental_tracks.song_title)
plt.show()
```
